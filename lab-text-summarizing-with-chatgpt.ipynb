# Text summarizing with ChaptGPT
In this lesson, you will summarize text with a focus on specific topics.

## Setup

````python
from openai import OpenAI
import os

from dotenv import load_dotenv, find_dotenv
_ = load_dotenv(find_dotenv()) # read local .env file

OPENAI_API_KEY  = os.getenv('OPENAI_API_KEY')
````

````python
client = OpenAI(
    # This is the default and can be omitted
    api_key=OPENAI_API_KEY,
)


def get_completion(prompt, model="gpt-3.5-turbo"): # Andrew mentioned that the prompt/ completion paradigm is preferable for this class
    messages = [{"role": "user", "content": prompt}]
    response = client.chat.completions.create(
        model=model,
        messages=messages,
        temperature=0, # this is the degree of randomness of the model's output
    )
    return response.choices[0].message.content
````

## Text to summarize

````python
prod_review = """
Got this panda plush toy for my daughter's birthday, \
who loves it and takes it everywhere. It's soft and \ 
super cute, and its face has a friendly look. It's \ 
a bit small for what I paid though. I think there \ 
might be other options that are bigger for the \ 
same price. It arrived a day earlier than expected, \ 
so I got to play with it myself before I gave it \ 
to her.
"""
````

## Summarize with a word/sentence/character limit

````python
prompt = f"""
Your task is to generate a short summary of a product \
review from an ecommerce site. 

Summarize the review below, delimited by triple 
backticks, in at most 30 words. 

Review: ```{prod_review}```
"""

response = get_completion(prompt)
print(response)
````

````python
prompt = f"""
Your task is to generate a short summary of a product \
review from an ecommerce site to give feedback to the \
Shipping deparmtment. 

Summarize the review below, delimited by triple 
backticks, in at most 30 words, and focusing on any aspects \
that mention shipping and delivery of the product. 

Review: ```{prod_review}```
"""

response = get_completion(prompt)
print(response)
````

### Exercise 3 — Laundry Service Operational Extraction

## Conclusions

Across the three exercises, the model consistently demonstrated its ability to summarize and extract information with a clear, department‑focused perspective. Each prompt required the model to ignore irrelevant details and concentrate only on the aspects specified—food quality in the arepera review, service quality in the dental clinic review, and operational details in the laundry service review. The outputs were accurate, concise, and aligned with the intended focus, showing that well‑designed prompts reliably guide the model toward the desired information.

The results also highlight an important lesson: when the instructions are specific, structured, and include clear constraints (such as word limits and defined focus areas), the model performs more precisely and avoids unnecessary or unrelated content. In all three cases, the summaries remained faithful to the original reviews without hallucinating or inventing details. This reinforces the value of targeted prompting for real‑world tasks like customer feedback analysis, departmental reporting, and operational monitoring.

Overall, the exercises demonstrate how effective prompt design can transform raw customer reviews into actionable insights, and how models can adapt their output depending on the analytical lens requested.
